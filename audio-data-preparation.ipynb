```python
# Extraction of the data from the mp4 files and saving them in the audio_files directory
# NOT TO BE RUN AGAIN

video_dir = 'data/Audio/Videos/'
audio_dir = 'data/Audio/AllAudio'

video_files = [f for f in os.listdir(video_dir) if f.endswith('_u.mp4')]

for video_file in video_files:
    clip = VideoFileClip(os.path.join(video_dir, video_file))
    audio_file = video_file.replace('_u.mp4', '.wav')
    clip.audio.write_audiofile(os.path.join(audio_dir, audio_file))
```

## Embedding Creation

```python
import os
import soundfile as sf

audio_dir = 'data/Audio/F/train/'
file_name = '1_S10E04_007.wav'  # Replace with an actual file name from your directory
file_path = os.path.join(audio_dir, file_name)

if not os.path.exists(file_path):
    raise FileNotFoundError(f"Audio file {file_path} does not exist.")

# Try loading the audio file using soundfile
try:
    waveform, sample_rate = sf.read(file_path)
    print(f"Loaded {file_name} successfully with sample rate {sample_rate} and waveform shape {waveform.shape}")
except Exception as e:
    print(f"Error loading {file_name}: {e}")
```

```python
audio_dir = 'data/Audio/F/train/'
embeddings_dir = 'data/Audio/F_embeddings/train/'


if not os.path.exists(audio_dir):
    raise FileNotFoundError(f"Audio directory {audio_dir} does not exist.")
if not os.path.exists(embeddings_dir):
    os.makedirs(embeddings_dir)
    
bundle = torchaudio.pipelines.WAV2VEC2_ASR_BASE_960H
model = bundle.get_model()

for file_name in os.listdir(audio_dir):
    if file_name.endswith('.wav'):
        file_path = os.path.join(audio_dir, file_name)
        print(file_path)
        y, sr = torchaudio.load(file_path)
        with torch.inference_mode():
            feats, _ = model.extract_features(torch.unsqueeze(y[0], dim=0))
            embeddings_path = os.path.join(embeddings_dir, file_name.replace('.wav', '.npy'))
            np.save(embeddings_path, feats[0].detach().numpy())
```
